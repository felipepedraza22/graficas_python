# GRAFICAS CON LA LIBRERIA MATPLOTLIB
## GRAFICAS EN DOS DIMENSIONES 

Gr√°ficos de una lista

```python
# importamos la libreria
#import matplolib.pyplot

from matplotlib.pyplot import plot, show, xlim, ylim, xticks
```

```python
# Grafica No. 1

y = [1,2,3,4,5]

plot(y)
show()
```

```python
# Grafica No. 2

y = [1,2,3,4,5]

plot(y, "*")
show()
```

```python
# Grafica No. 3

y = [1,2,3,4,5]
x = [2,4,6,8,10]

plot(x, y, "*")
show()
```

```python
# Frafica No. 4

# Grafica con limites

y = [1,2,3,4,5]
x = [2,4,6,8,10]

plot(x, y, "*")
xlim(-10, 10)
ylim(0,15)
show()
```

```python
#Grafica No. 5

from numpy import linspace, pi, sin

x= linspace(-10*pi, 10*pi, 2560)
y= sin(x)

plot(x,y)
ylim(-2,2)
show()
```

```python
#Grafica No.6
from numpy import linspace, pi, sin

x= linspace(-pi, pi, 256)
y= sin(x)

plot(x,y)
ylim(-2,2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
show()
```
